```python
m=[]
y=[]
for i in df4['Formatted Date']:
    split_date = i.split('-')
    number = split_date 
    year=number[0]
    y.append(year)
    month=number[1]
    day=number[2].split(' ')
    d=f'{month}/{day[0]}/{year}'
    m.append(d)  
df4['date']=m
```
